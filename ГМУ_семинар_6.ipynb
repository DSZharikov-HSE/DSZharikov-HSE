## Лямбда-функции

Лямбда-функции – это анонимные функции, которые могут включать только одно выражение.

Они обычно используются для выполнения простых операций, не требующих полного определения функции с помощью ключевого слова `def`. Лямбда-функции могут принимать любое количество аргументов, но могут возвращать только одно значение.

Синтаксис:

`lambda argument1, argument2, ...: expression`

## Мотивация

При обработке данных часто возникает необходимость использовать одни и те же методы несколько раз.

Пример:

*Анализ финансовых показателей в различных регионах*

```python
numbers = [54, 26, 87, 83, 29, 56, 21, 8, 16, 94, 34, 86, 13, 41, 97, 74, 81, 52, 3, 82]

# вычисление медианы
numbers.sort()
n = len(numbers)
if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
    median = numbers[n//2]
print("Медиана:", median)

# вычисление моды
mode_dict = {}
for num in numbers:
    if num in mode_dict:
        mode_dict[num] += 1
    else:
        mode_dict[num] = 1
mode = max(mode_dict, key=mode_dict.get)
print("Мода:", mode)

# вычисление среднего
mean = sum(numbers) / n
print("Среднее:", mean)

# вычисление дисперсии
mean_diff = sum((x - mean) ** 2 for x in numbers)
variance = mean_diff / n
print("Дисперсия:", variance)
```

*В данном регионе наблюдается ...*

```python
numbers = [48, 11, 8, 98, 17, 94, 63, 95, 97, 44, 41, 42, 96, 53, 15, 33, 20, 76, 21, 46]

# вычисление медианы
numbers.sort()
n = len(numbers)
if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
    median = numbers[n//2]
print("Медиана:", median)

# вычисление моды
mode_dict = {}
for num in numbers:
    if num in mode_dict:
        mode_dict[num] += 1
    else:
        mode_dict[num] = 1
mode = max(mode_dict, key=mode_dict.get)
print("Мода:", mode)

# вычисление среднего
mean = sum(numbers) / n
print("Среднее:", mean)

# вычисление дисперсии
mean_diff = sum((x - mean) ** 2 for x in numbers)
variance = mean_diff / n
print("Дисперсия:", variance)
```

*В данном регионе наблюдается ... По итогу сравнения можно сказать ...*

Переиспользуем один и тот же код несколько раз. Программу неудобно читать, возникает нагромождение кода. А что будет, если понадобится исследовать 100, 1000 населенных пунктов? Решение - функции!

## Основная информация про функции

Функции в Python — это один из основных инструментов для создания удобного и структурированного кода. Они позволяют группировать набор инструкций, которые выполняют определенную задачу, и использовать их многократно в вашей программе.

Для создания функции в Python используется ключевое слово `def`, за которым следует имя функции и круглые скобки `()`.
В скобках указываются аргументы функции (если они есть), а после скобок ставится двоеточие `:`. Тело функции записывается с отступом.

Количество и тип переменных может быть любым.



Синтаксис:

```python
def function_name(arg1, arg2, ...):
    function_body
```

Пример простой функции без аргументов:

```python
def greet():
    print("Hello, world!")
```

```python
# вызов функции
greet()
```

## `return`

Функции могут возвращать значение с помощью ключевого слова return. Значение, следующее после return, будет результатом работы функции.

Если функция возвращает несколько значений, то результат работы функции - кортеж с возвращаемыми параметрами.

Синтаксис:

```python
def func(arg1, arg2, arg3, arg4):
    body_function
    return result1, result2, result3
```

```python
def add(a, b):
    return a + b

result = add(2, 3)
print(result)
```

```python
def add_c(a, b, c):
    return a + b, a + c

result = add_c(2, 3, 4)
print(result)
```

Пример из начала семинара:

```python
def make_eda(numbers):
    # вычисление медианы
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        median = (numbers[n//2 - 1] + numbers[n//2]) / 2
    else:
        median = numbers[n//2]
    # вычисление моды
    mode_dict = {}
    for num in numbers:
        if num in mode_dict:
            mode_dict[num] += 1
        else:
            mode_dict[num] = 1
    mode = max(mode_dict, key=mode_dict.get)
    # вычисление среднего
    mean = sum(numbers) / n
    # вычисление дисперсии
    mean_diff = sum((x - mean) ** 2 for x in numbers)
    variance = mean_diff / n
    return median, mode, mean, variance
```

```python
numbers = [3, 8, 13, 16, 21, 26, 29, 34, 41, 52, 54, 56, 74, 81, 82, 83, 86, 87, 94, 97]
```

```python
median_val, mode_val, mean_val, variance_val = make_eda(numbers)
print(median_val, mode_val, mean_val, variance_val)
```

## Ключевое слово `global`

Для изменения переменной объявленной вне функции используется ключевое слово `global`.

```python
a = 4

def func():
  a += 3
  print(a)

func()
print(a)
```

```python
a = 4

def func():
  global a
  a += 3
  print(a)

func()
print(a)
```
