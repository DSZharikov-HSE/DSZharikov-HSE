## Операторы сравнения

Операторы сравнения используются для сравнения значения и возвращают логический тип. Таких операторов несоклько:
* == (равно)
* != (не равно)
* \> (больше чем)
* < (меньше чем)
* \>= (больше или равно)
* <= (меньше или равно)

```python
income = 15000
minimum_wage = 15000
is_minimum_wage = income == minimum_wage
print("Is the income equal to the minimum wage?:", is_minimum_wage)
```

```python
age = 25
retirement_age = 65
is_retirement_eligible = age < retirement_age
print("Is the person eligible for retirement?:", is_retirement_eligible)
```

```python
votes = 1000
votes_needed = 1000
is_winning = votes >= votes_needed
print("Is the candidate winning?:", is_winning)
```

```python
income = 50000
tax_bracket_lower = 15000
tax_bracket_upper = 30000
is_in_tax_bracket = tax_bracket_lower <= income <= tax_bracket_upper # Внимание на интервалы!
print("Is the income in the tax bracket?:", is_in_tax_bracket)
```

```python
tax_bracket_lower <= income <= tax_bracket_upper
```

## Логические операторы

Логические операторы используются для комбинирования логических типов

В питоне три логических оператора: and, or и not

and возвращает True, только если оба значения True

or возваращет True, если хотя бы одно значение True

not возвращает противоположное значение

image.png

```python
income = 50000
minimum_wage = 15000
is_minimum_wage = income >= minimum_wage
print(is_minimum_wage)
is_taxable = income >= 50000
print(is_taxable)
is_minimum_wage_and_taxable = is_minimum_wage and is_taxable
print("Is the income both minimum wage and taxable?:", is_minimum_wage_and_taxable)
```

```python
is_registered = True
is_voted = False
did_participate = is_registered and is_voted
print("Did the person participate in the election?:", did_participate)
```

```python
is_minimum_wage = True
is_above_minimum_wage = not is_minimum_wage
print("Is the income above the minimum wage?:", is_above_minimum_wage)
```

## Ветвление (if-elif-else)

image.png

```python
income = 25000

if income <= 15000:
    tax_bracket = "Low"
elif (income > 15000) and (income < 20000):
    tax_bracket = "Moderate"
else:
    tax_bracket = "High"
```

```python
income = 25000

if income <= 15000:
    tax_bracket = "Low"
elif (income > 15000) and (income < 20000):
    tax_bracket = "Moderate"
else:
    tax_bracket = "High"

if tax_bracket == "Low":
    tax_rate = 0.1
elif tax_bracket == "Moderate":
    tax_rate = 0.2
else:
    tax_rate = 0.3

tax_owed = income * tax_rate

print(f"Income: ${income}")
print(f"Tax Bracket: {tax_bracket}")
print(f"Tax Rate: {tax_rate}")
print(f"Tax Owed: ${tax_owed}")
```

```python
income = 50000
tax_bracket_lower = 15000
tax_bracket_upper = 30000
is_taxable = tax_bracket_lower <= income <= tax_bracket_upper
if not is_taxable:
    print("The income is not taxable.")
else:
    print("The income is taxable.")
```

```python
income = 50000
tax_bracket_lower = 15000
tax_bracket_upper = 30000
is_taxable = tax_bracket_lower <= income <= tax_bracket_upper
if is_taxable:
    print("The income is taxable.")
else:
    print("The income is not taxable.")
```

```python
income = 50000
minimum_wage = 15000
tax_bracket_lower = 15000
tax_bracket_upper = 30000
is_minimum_wage = income >= minimum_wage
is_taxable = tax_bracket_lower <= income <= tax_bracket_upper
if is_minimum_wage and is_taxable:
    print("The income is both above minimum wage and taxable.")
elif is_minimum_wage or is_taxable:
    print("The income is either above minimum wage or taxable.")
else:
    print("The income is neither above minimum wage nor taxable.")
```

## Истина и ложь

В питоне некоторые объекты называются truthy, а некоторые falsy.

Truthy - это те, которые при преобразовании в логический тип, становятся True, falsy - в false

Примеры truthy значений: True, любое ненулевое число, любая непустая строка, любой непустой список
Примеры. falsy значений: False, None, 0, "", []
